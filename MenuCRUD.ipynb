Una vez visto el ***Módulo 1: Fundamentos de Programación***, se presenta la siguiente actividad para que el alumnado pueda poner a prueba lo aprendido durante este módulo.

Dentro de la siguiente dinámica, realizar un **menú CRUD** donde el tema a emplear es libre. Cada función deberá ejecutar una **operación del CRUD**, agregando una opción de **visualización**:

*   **Ver todos los registros**
*   **Crear nuevo registro**
*   **Consultar registro**
*   **Actualizar registro**
*   **Eliminar registro**
*   **Salir menú**

```python
experimentos = [
    {
        "Nombre": "Doble Rendija",
        "Investigador": "Thomas Young",
        "Area": "Física",
        "Año": 1801,
        "Resultado": "Demostró la naturaleza ondulatoria de la luz"
    },
    {
        "Nombre": "Penicilina",
        "Investigador": "Alexander Fleming",
        "Area": "Biología",
        "Año": 1928,
        "Resultado": "Descubrimiento del primer antibiótico"
    },
    {
        "Nombre": "Relatividad Especial",
        "Investigador": "Albert Einstein",
        "Area": "Física",
        "Año": 1905,
        "Resultado": "Dilatación del tiempo y espacio"

    },
    {
        "Nombre": "Teoría endosimbiótica",
        "Investigador": "Lynn Margulis",
        "Area": "Biología",
        "Año": 1967,
        "Resultado": "Evolución de células eucariotas a partir de procariotas"

    }
]
```

```python
def ver_experimentos(experimentos):

    if not experimentos:
        print("No hay experimentos registrados.")
    else:
        for i, exp in enumerate(experimentos):
            print(f"\n========= Experimento {i+1} ===========")
            print(f"Nombre: {exp['Nombre']}")
            print(f"Investigador: {exp['Investigador']}")
            print(f"Área: {exp['Area']}")
            print(f"Año: {exp['Año']}")
            print(f"Resultado: {exp['Resultado']}")
```

```python
ver_experimentos(experimentos)
```

```python
def crear_experimento(experimentos):
    nombre = input("Nombre del experimento: ")
    investigador = input("Investigador: ")
    area = input("Área científica: ")
    año = input("Año: ")
    resultado = input("Resultado principal: ")

    nuevo = {
        "Nombre": nombre,
        "Investigador": investigador,
        "Area": area,
        "Año": año,
        "Resultado": resultado
    }

    experimentos.append(nuevo)
    print("\nExperimento registrado con éxito.")
```

```python
crear_experimento(experimentos)
```

Nombre del experimento: Experimento de Oersted
Investigador: Hans Christian Oersted
Área científica: Física
Año: 1820
Resultado principal: Relación entre la electricidad y el magnetismo

Experimento registrado con éxito

```python
def buscar_experimento(experimentos, accion):

    nombres = []
    for exp in experimentos:
        nombres.append(exp["Nombre"])

    print("----Estos son los experimentos:\n")
    print("\n".join(nombres))
    print()

    a = input(f"¿Qué experimento quieres {accion}? \n")
    print()
    for exp in experimentos:
        if exp["Nombre"].lower() == a.lower():
            return exp

    return None
```

```python
buscar_experimento(experimentos, "ver")
```

```python
def actualizar_experimento(experimentos):

    while True:

        exp = buscar_experimento(experimentos, "actualizar")

        if exp is None:
            print("Experimento no encontrado.")
            return

        while True:
            print("===== ACTUALIZAR REGISTRO =====")
            print("1. Actualizar nombre")
            print("2. Actualizar investigador")
            print("3. Actualizar área")
            print("4. Actualizar año")
            print("5. Actualizar resultado")
            print("6. Elegir otro experimento")
            print("7. Salir")
            print()
            opcion = input("Selecciona una opción: ")
            print()

            if opcion == "1":
                exp["Nombre"] = input("Nuevo nombre: ")
            elif opcion == "2":
                exp["Investigador"] = input("Nuevo investigador: ")
            elif opcion == "3":
                exp["Area"] = input("Nueva área: ")
            elif opcion == "4":
                exp["Año"] = input("Nuevo año: ")
            elif opcion == "5":
                exp["Resultado"] = input("Nuevo resultado: ")
            elif opcion == "6":
                break
            elif opcion == "7":
                return
            else:
                print("Opción no válida.")

            print("\nRegistro actualizado:", exp)
            print()
```

```python
actualizar_experimento(experimentos)
```

```python
def eliminar_experimento(experimentos):

    nombres = []
    for exp in experimentos:
        nombres.append(exp["Nombre"])

    print("----Estos son los experimentos:\n")
    print("\n".join(nombres))
    print()

    nombre = input("Nombre del experimento a eliminar: ")

    for exp in experimentos:
        if exp["Nombre"].lower() == nombre.lower():
            experimentos.remove(exp)
            print("Experimento eliminado correctamente.")
            return

    print("Experimento no encontrado.")
```

```python
eliminar_experimento(experimentos)
```

```python
def menu_experimentos(experimentos):
    while True:
        print("\n===== SISTEMA DE GESTIÓN DE EXPERIMENTOS =====")
        print()
        print("1. Ver todos los registros")
        print("2. Crear nuevo registro")
        print("3. Consultar registro")
        print("4. Actualizar registro")
        print("5. Eliminar registro")
        print("6. Salir")
        print()
        opcion = input("Selecciona una opción: ")
        print()
        if opcion == "1":
            ver_experimentos(experimentos)
        elif opcion == "2":
            crear_experimento(experimentos)
        elif opcion == "3":
            buscar_experimento(experimentos, accion="consultar")
        elif opcion == "4":
            actualizar_experimento(experimentos)
        elif opcion == "5":
            eliminar_experimento(experimentos)
        elif opcion == "6":
            print("Saliendo, gracias por consultar los experimentos.")
            break
        else:
            print("Opción no encontrada.")
```

```python
menu_experimentos(experimentos)
```
